```python
dp= datap.copy() 
```

```python
X = dp.drop(['Outcome'], axis=1 )
    
```

```python
X = X.drop(['DiabetesPedigreeFunction'], axis=1 )
```

```python
X
```

```python
y= dp['Outcome']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.2, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,y_train)
```

```python
x_train_pred = classifier.predict(X_train)
training_data_accuracy = accuracy_score(x_train_pred,y_train)
print(training_data_accuracy)
```

```python
#accuracy score on the test data
x_test_pred = classifier.predict(X_test)
test_data_accuracy = accuracy_score(x_test_pred,y_test)
print('Accuracy score of the test data : ',test_data_accuracy)
```

```python
input_data = [5,174,85,23,163,25.8,78]

#changing array to numpy array
numpy_array = np.asarray(input_data)

#reshape the array
reshaped_array = numpy_array.reshape(1,-1)

#standardise the input data
std_data = scaler.transform(reshaped_array)
print(std_data)

prediction  = classifier.predict(std_data)
print(prediction)

if prediction == 1:
    print('diabetic')
else:
    print('Non diabetic')
```
